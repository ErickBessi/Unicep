## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

### Exercício 3: Validação de Formato de Placa de Carro (Mercosul)
Desenvolva um programa que valide se uma string representa uma placa de carro no formato Mercosul. O formato é `LLLNLNN` (três letras, um número, uma letra, dois números) OU `LLLNNLN` (três letras, dois números, uma letra, um número).

Considere as seguintes condições:
- O comprimento da string deve ser 7.
- As posições das letras devem ser caracteres alfabéticos.
- As posições dos números devem ser dígitos numéricos.
- Faça a validação usando condicionais aninhadas e operadores lógicos para verificar cada posição da string.

```python
placa = input("Digite a placa do carro:").lower()

# 1. Verifica o comprimento da placa primeiro para evitar IndexError
if len(placa) != 7:
  print("Erro: A placa Mercosul deve ter 7 caracteres.")
else:
  # Formato LLLNLNN (Ex: ABC1D23)
  condicao1 = (
      placa[0].isalpha() and
      placa[1].isalpha() and
      placa[2].isalpha() and
      placa[3].isdigit() and
      placa[4].isalpha() and
      placa[5].isdigit() and
      placa[6].isdigit()
  )

  # Formato LLLNNLN (Ex: ABC12D3)
  condicao2 = (
      placa[0].isalpha() and
      placa[1].isalpha() and
      placa[2].isalpha() and
      placa[3].isdigit() and
      placa[4].isdigit() and
      placa[5].isalpha() and
      placa[6].isdigit()
  )

  if condicao1 or condicao2:
    print("Placa aceita: Formato Mercosul válido!")
  else:
    print("Erro: Formato de placa Mercosul inválido. Redigite a placa.")
```

Você tem um pacote para entregar e precisa escolher o veículo adequado com base no peso do pacote.

Peça:
- Peso do Pacote

**Regras de Seleção de Veículo:**
- Se o peso do pacote for menor ou igual a 10 kg, o veículo ideal é uma 'Moto'.
- Se o peso do pacote for maior que 10 kg e menor ou igual a 50 kg, o veículo ideal é uma 'Van Pequena'.
- Se o peso do pacote for maior que 50 kg e menor ou igual a 200 kg, o veículo ideal é um 'Caminhão Pequeno'.
- Se o peso do pacote for maior que 200 kg, é necessário um 'Caminhão Grande'.

Determine e informe qual o veículo mais adequado para a entrega do pacote.

```python
pesoPacote= int(input("Digite o peso do pacote em KG: "))
if (pesoPacote<=10):
  print("Veiculo ideal: Moto")
elif (pesoPacote>=10 and pesoPacote<=50):
  print("Veiculo ideal: Van Pequena")
elif (pesoPacote>=50 and pesoPacote<=200):
  print("Veiculo ideal: Caminhão Pequeno")
elif (pesoPacote>200):
  print("Veiculo ideal: Caminhão Grande")
```
