# Aula 10: Arquivos JSON e CSV

Até agora, toda vez que o nosso programa terminava, todas as nossas variáveis (listas, dicionários) eram perdidas.
Para salvar os dados permanentemente (persistência), precisamos aprender a ler e escrever em **Arquivos**.

## 1. Arquivos de Texto Simples `.txt`

O Python possui a função `open()` para abrir arquivos. Precisamos dizer o nome do arquivo e o **Modo** de abertura:
- `'r'` (read): Apenas para Ler (padrão).
- `'w'` (write): Escrever por cima de tudo.
- `'a'` (append): Escrever no final do arquivo.

```python
# Escrevendo em um arquivo (vai ser criado nesta mesma pasta)
arquivo = open("meu_diario.txt", "w")                     # O \n é usado para pular linha
arquivo.write("Hoje comecei a aprender sobre arquivos em Python.\n")
arquivo.close() # SEMPRE feche o arquivo quando terminar!
```

**Lembrar de:**
- Ls, ver o que está na pasta
- Ls -la, Ver todos os arquivos
- Cat nomeDoArquivo, ver o conteúdo do arquivo
(Aprender Linux)


### O bloco `with`
Esquecer o `arquivo.close()` é comum e pode corromper os dados. A forma profissional de fazer isso no Python é usando o bloco `with`. Ele abre e **garante que vai fechar**.

```python
# Lendo o arquivo usando with (Muito mais seguro e limpo!)
with open("meu_diario.txt", "r") as arquivo:
    conteudo = arquivo.read()
    print("Li do arquivo o seguinte texto:\n" + conteudo)

# Já está fechado aqui fora!
```

## 2. Arquivos JSON

O `.txt` é muito básico. E se quisermos salvar um Dicionário ou uma Lista complexa?
O formto **JSON** (JavaScript Object Notation) é o padrão de mercado para isso na internet. Ele é basicamente texto, mas formatado exatamente como os Dicionários e Listas do Python!

```python
import json

meus_dados = {
    "nome": "Saulo",
    "habilidades": ["Python", "Dados"],
    "nivel": 10
}

# Salvando um Dicionário Python para um arquivo JSON (DUMP)
with open("perfil.json", "w") as f:
    json.dump(meus_dados, f, indent=4)

print("Arquivo JSON criado! Abra a pasta e dê um duplo clique no perfil.json para olhar.")
```

```python
# Lendo um JSON devolta para o Python (LOAD)
with open("perfil.json", "r") as f:
    dados_carregados = json.load(f)

print(f"Bem-vindo de volta, {dados_carregados['nome']}!")
```

## 3. Arquivos CSV (Excel / Planilhas)

CSV (Comma Separated Values) é o formato padrão para exportar tabelas. São como planilhas do Excel, onde cada linha do arquivo é uma linha da tabela, e as colunas são separadas por vírgula `,` ou ponto-e-vírgula `;`.
O Python possui o módulo nativo `csv` para ajudar.

```python
import csv

# Escrevendo dados tabulares
alunos = [
    ["Nome", "Idade", "Nota"],
    ["Maria", 22, 9.5],
    ["José", 25, 7.0]
]

with open("turma.csv", "w", newline='', encoding='utf-8') as f:
    escritor = csv.writer(f)
    escritor.writerows(alunos)

print("Arquivo da turma criado!")
```

## Exercício Resolvido: Sistema de Cadastro em Arquivo Texto

A seguir, um exemplo de um sistema completo de cadastro que utiliza arquivos de texto (`.txt`) para persistência de dados. Este sistema permite incluir, listar, alterar e eliminar logicamente registros.

```python
# ============================================
# Sistema de Cadastro em Arquivo Texto
# ============================================

ARQUIVO = "dados.txt"


# --------------------------------------------
# Função para incluir registro
# --------------------------------------------
def incluir():
    codigo = input("Código: ")
    nome = input("Nome: ")
    idade = input("Idade: ")

    flag = "True"

    registro = f"{codigo};{nome};{idade};{flag}\n"

    with open(ARQUIVO, "a", encoding="utf-8") as f:
        f.write(registro)

    print("Registro incluído com sucesso!\n")


# --------------------------------------------
# Função para listar registros válidos
# --------------------------------------------
def listar():
    try:
        with open(ARQUIVO, "r", encoding="utf-8") as f:
            linhas = f.readlines()

        print("\n===== REGISTROS VÁLIDOS =====")

        encontrou = False

        for linha in linhas:
            dados = linha.strip().split(";")

            codigo = dados[0]
            nome = dados[1]
            idade = dados[2]
            flag = dados[3]

            if flag == "True":
                encontrou = True
                print(f"Código: {codigo}")
                print(f"Nome..: {nome}")
                print(f"Idade.: {idade}")
                print("----------------------")

        if not encontrou:
            print("Nenhum registro válido encontrado.")

    except FileNotFoundError:
        print("Arquivo ainda não existe.")

    print()


# --------------------------------------------
# Função para alterar registro
# --------------------------------------------
def alterar():
    codigo_busca = input("Digite o código a alterar: ")

    try:
        with open(ARQUIVO, "r", encoding="utf-8") as f:
            linhas = f.readlines()

        novo_conteudo = []
        encontrado = False

        for linha in linhas:
            dados = linha.strip().split(";")

            codigo = dados[0]
            nome = dados[1]
            idade = dados[2]
            flag = dados[3]

            if codigo == codigo_busca and flag == "True":
                print("Registro encontrado!")

                novo_nome = input("Novo nome: ")
                nova_idade = input("Nova idade: ")

                nova_linha = f"{codigo};{novo_nome};{nova_idade};True\n"

                novo_conteudo.append(nova_linha)

                encontrado = True
            else:
                novo_conteudo.append(linha)

        with open(ARQUIVO, "w", encoding="utf-8") as f:
            f.writelines(novo_conteudo)

        if encontrado:
            print("Registro alterado com sucesso!\n")
        else:
            print("Registro não encontrado.\n")

    except FileNotFoundError:
        print("Arquivo não encontrado.\n")


# --------------------------------------------
# Função para eliminação lógica
# --------------------------------------------
def eliminar():
    codigo_busca = input("Digite o código a eliminar: ")

    try:
        with open(ARQUIVO, "r", encoding="utf-8") as f:
            linhas = f.readlines()

        novo_conteudo = []
        encontrado = False

        for linha in linhas:
            dados = linha.strip().split(";")

            codigo = dados[0]
            nome = dados[1]
            idade = dados[2]
            flag = dados[3]

            if codigo == codigo_busca and flag == "True":
                # Eliminação lógica
                nova_linha = f"{codigo};{nome};{idade};False\n"

                novo_conteudo.append(nova_linha)

                encontrado = True
            else:
                novo_conteudo.append(linha)

        with open(ARQUIVO, "w", encoding="utf-8") as f:
            f.writelines(novo_conteudo)

        if encontrado:
            print("Registro eliminado logicamente.\n")
        else:
            print("Registro não encontrado.\n")

    except FileNotFoundError:
        print("Arquivo não encontrado.\n")


# --------------------------------------------
# Menu principal
# --------------------------------------------
def menu():
    while True:
        print("========== MENU ==========")
        print("1 - Incluir")
        print("2 - Listar")
        print("3 - Alterar")
        print("4 - Eliminar Logicamente")
        print("5 - Sair")
        print("========================")

        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            incluir()

        elif opcao == "2":
            listar()

        elif opcao == "3":
            alterar()

        elif opcao == "4":
            eliminar()

        elif opcao == "5":
            print("Encerrando programa...")
            break

        else:
            print("Opção inválida!\n")


# --------------------------------------------
# Programa principal
# --------------------------------------------
menu()
```

### Desafio 1: Agenda Telefônica JSON
Crie um dicionário vazio `agenda = {}`. Peça ao usuário em um loop `while` o nome e o telefone, e os guarde na agenda até ele digitar "sair". No final, use `json.dump` para salvar a agenda em um arquivo `agenda.json`.

```python
import json

agenda = {}

# Seu código aqui

while True:
  print("=======================")
  print("1- Adicionar número de telefone")
  print("2- Sair")
  menu= int(input("Digite uma das opções: "))
  print("=======================")
  if menu==1:
    nomeDigitado= input("Digite o nome: ")
    telDigitado= input("Digit o número de telefone: ")
    agenda['nome']= nomeDigitado
    agenda['tel']= telDigitado
  elif menu==2:
    break

# Salvando um Dicionário Python para um arquivo JSON (DUMP)
with open("agenda.json", "w") as arquivo:
    json.dump(agenda, arquivo, indent=4)

print("Arquivo JSON criado com sucesso.")
```

### Desafio 2: Cadastro de Produtos em CSV
Foi te pedida a criação de um sisteminha básico de controle de estoque.
Crie uma lista de listas chamada `produtos`, por exemplo: `[["Notebook", 3500.0, 5], ["Mouse", 120.0, 15]]`.
Abra um arquivo chamado `estoque.csv` em modo de escrita (`'w'`) e use o módulo `csv.writer` para escrever o cabeçalho `["Produto", "Preco", "Quantidade"]` e logo após, as linhas dos produtos. Ao final, confira se o arquivo foi criado com as vírgulas corretamente.

```python
import csv
estoque=[
    ["Notebook",3500,5],
    ["Mouse",120,15],
    ["Fone de ouvido",150,10]
]
with open("estoque.csv", "w", newline='') as arquivo:
  writer = csv.writer(arquivo)
  writer.writerow(["Produto", "Preco", "Quantidade"])
  writer.writerows(estoque)


```
