## Exemplo Resolvido 1: Tabuada

**O problema:** Crie um programa que peça um número ao usuário e imprima a tabuada desse número de 1 a 10 usando o `for`.

```python
numero = int(input("Você quer a tabuada de qual número? "))

print(f"\n--- Tabuada do {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2d} = {resultado:2d}")
```

*Dica: O formato `{:2d}` dentro da f-string diz para o Python reservar 2 espaços na hora de imprimir aquele número. Isso ajuda a deixar os números alinhados bonitinhos na tela!*

## Exemplo Resolvido 2: Somatória com While

**O problema:** Peça números ao usuário repetidamente. O programa deve somar todos os números digitados e só deve parar (e mostrar o total) quando o usuário digitar o **número ZERO**.

```python
soma = 0
while True:
    valor = float(input("Digite um valor para somar (ou 0 para parar): "))

    if valor == 0:
        break # Sai do loop quando o usuário digitar zero

    soma = soma + valor

print(f"\nA soma total foi de: {soma}")
```

### Desafio 21.
Faça um programa que peça um número inteiro e determine se ele é ou não um número primo. Um número primo é aquele que é divisível somente por ele mesmo e por 1.

```python
entrada = int(input("Digite um valor: "))

divisores = []
for n in range(1, entrada + 1):
    if entrada % n == 0:
        divisores.append(n)

if len(divisores) == 2:
    print(f"{entrada} é primo")
else:
    print(f"{entrada} não é primo")
```

### Desafio 23.
Faça um programa que mostre todos os primos entre 1 e N sendo N um número inteiro fornecido pelo usuário. O programa deverá mostrar também o número de divisões que ele executou para encontrar os números primos. Serão avaliados o funcionamento, o estilo e o número de testes (divisões) executados.

```python
# Escreva seu código aqui
entrada = int(input("Digite um valor: "))
cont=0
nPrimos=[]
nAtual=2
#Rodar do número 2(primeiro número primo) até o número que o usuário digitar
while nAtual<=entrada:
  #divisores dentro do while para cada interação ele ser limpo
  divisores = []
  #Laço de repetição for para verificar se um número é primo
  #Número atual serve para verificar todos os números primos até o número que o usuário digitar
  for n in range(1, nAtual + 1):
      if nAtual % n == 0:
          divisores.append(n)
          cont+=1
  if len(divisores) == 2:
    nPrimos.append(nAtual)

  nAtual+=1
print(f"Número de divisões executadas: {cont}")
print(f"Número digitado:{entrada}, de um até o número digitado existem os seguintes números primos:{nPrimos}", end=" ")
```

### Desafio 26.
Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
# Escreva seu código aqui
votosA=[]
votosB=[]
votosC=[]
eleitores= int(input("Digite a quantidade total de eleitores: "))
for n in range (1,eleitores+1):
  print("Nesta votação temos os Canditatos A,B e C")
  voto= input("Qual o seu voto? ").lower()
  if voto=="a":
    votosA.append(1)
  elif voto=="b":
    votosB.append(1)
  elif voto=="c":
    votosC.append(1)
  else:
    print("Voto inválido")
if (sum(votosA)>sum(votosB) and sum(votosA)>sum(votosC)):
  print(f"Candidato A ganhou com {sum(votosA)} votos")
elif (sum(votosB)>sum(votosC)):
  print(f"Candidato B ganhou com {sum(votosB)} votos")
else:
  print(f"Candidato C ganhou com {sum(votosC)} votos")
```
