# Aula 07: Listas

Até agora, nossas variáveis (`idade`, `nome`, `preco`) guardavam **um único valor** por vez.
Mas e se precisássemos guardar a nota de 50 alunos? Criar 50 variáveis (`nota1`, `nota2`, ..., `nota50`) seria terrível.
Para guardar uma coleção de valores em uma única variável, usamos as **Listas**.

## 1. Criando Listas

Listas são criadas usando colchetes `[]` e separando os itens por vírgulas.
Podem conter números, strings, misturas, e até outras listas!

```python
notas = [8.5, 7.0, 9.2, 5.5, 10.0]
frutas = ["Maçã", "Banana", "Laranja"]
lista_mista = ["João", 25, True, 1.75]

print("Notas:", notas)
print("Frutas:", frutas)
```

## 2. Acessando Itens da Lista

Cada item na lista tem um "endereço" (posição/índice). O Python **começa a contar do 0**.
Isso significa que o 1º item está no índice 0, o 2º no índice 1, etc.

```python
print("Primeira fruta:", frutas[0])
print("Segunda fruta:", frutas[1])

# O Python também permite o uso de índices negativos para contar de trás pra frente!
print("Última fruta:", frutas[-1])
```

## 3. Alterando e Adicionando Itens

Listas em Python são mutáveis, ou seja, podemos trocar, adicionar e remover itens delas quando quisermos.

```python
cesta = ["Pera", "Uva", "Maçã"]
print("Cesta original:", cesta)

# Alterando um item pelo índice
cesta[1] = "Morango"
print("Troquei Uva por Morango:", cesta)

# Adicionando um novo item NO FINAL da lista com .append()
cesta.append("Goiaba")
print("Adicionei Goiaba no final:", cesta)

# Removendo um item da lista com .remove()
cesta.remove("Pera")
print("Removi a Pera:", cesta)
```

## 4. O Poderoso Casamento: `for` + Listas

Listas foram feitas para serem usadas com o laço `for`. Veja como é fácil passar por todos os itens de uma lista:

```python
carrinho = ["Pão", "Leite", "Manteiga", "Café"]

print("Itens do meu carrinho:")
for item in carrinho:
    print(f"- Eu preciso comprar: {item}")
```

### Funções Úteis com Listas (Tamanho, Soma, Máximo)
O Python tem funções embutidas que nos ajudam a manipular dados em listas muito rapidamente.

```python
precos = [10.50, 4.20, 15.00, 2.50]

# len() para ver quantos itens existem na lista
print(f"Quantos itens? {len(precos)}")

# sum() para somar os números
# max() e min() para o maior e menor número
print(f"Total da compra: R$ {sum(precos)}")
print(f"Produto mais caro: R$ {max(precos)}")
print(f"Produto mais barato: R$ {min(precos)}")
```

### Desafio 6: Removendo Duplicatas
Dada uma lista com números repetidos, crie uma nova lista que contenha os mesmos números, mas sem duplicatas (cada número deve aparecer apenas uma vez). Dica: Pense em conjuntos ou em como iterar e verificar se o elemento já existe na nova lista.

```python
numeros1 = [1,1,2,2,2,3,4,4,4,4,4]
##criado a lista que vai armazenar a saida
numeros2 = []
#Verfica quantas repetições vai ter
qtd_numeros1 = len(numeros1)
#Faz uma varredura no tamanho da lista 1
for i in range(qtd_numeros1):
    #Flag é utilizado para verificar se  o numero na posição i vai ser adicionado ou não
    flag = True
    #Tamanho da lista 2
    qtd_numeros2 = len(numeros2)
    #Faz um for para repetir pela quantidade de números na lista 2
    for j in range(qtd_numeros2):
      #Compara as listas nas suas posições, se encontra algo ele não adiciona nada
        if numeros1[i] == numeros2[j]:
            flag = False
            break
    #Se foi encontrado um número novo, ele é adicionado
    if flag == True:
        numeros2.append(numeros1[i])
print(f"numeros2:",numeros2)
```

### Desafio 7: Média Ponderada (Listas Aninhadas)
Você tem uma lista de alunos, onde cada aluno é representado por uma sublista com `[nome, nota1, nota2, nota3]`. Calcule a média final de cada aluno (ponderando nota1 com peso 2, nota2 com peso 3 e nota3 com peso 5) e imprima o nome do aluno e sua média final.

```python
alunos = [
    ["Erick", 8, 9, 10],
    ["João", 10, 9, 8],
    ["Adrian", 8, 10, 9]
]

for pessoa in alunos:
    nome = pessoa[0]
    nota1 = pessoa[1]
    nota2 = pessoa[2]
    nota3 = pessoa[3]

    notaFinal = int((nota1 * 2 + nota2 * 3 + nota3 * 5) / (2 + 3 + 5))

    print(f"Nome: {nome}")
    print(f"Nota 1 = {nota1}")
    print(f"Nota 2 = {nota2}")
    print(f"Nota 3 = {nota3}")
    print(f"Média Final = {notaFinal}")
    print("-------------------")
```

### Desafio 9: Ordenando Objetos (Listas de Dicionários)

### Desafio 10: Análise de Estoque
Você tem uma lista de produtos em estoque, onde cada produto é uma sublista `[nome_produto, quantidade, preco_unitario]`. Crie um programa que faça o seguinte:
1. Calcule o valor total do estoque (quantidade * preco_unitario para cada produto e depois a soma de todos).
2. Encontre o produto com a maior quantidade em estoque.
3. Encontre o produto com o menor preço unitário.
Imprima todos os resultados de forma clara.

```python
# Seu código aqui
produtos=[
    ['PC',5,100],
    ['Celular',20,50],
    ['Console',11,200]
]

while True:
  print("=======================")
  print("1- Valor total do estoque")
  print("2- Produto com maior quantidade em estoque")
  print("3- Produto com menor preço unitário")
  print("4- Sair")
  menu= int(input("Qual opção você deseja ver? "))
  print("=======================")

  #Valor total em estoque
  if menu==1:
    somatorio=0
    for n in produtos:
      valorTotal= n[1]*n[2]
      somatorio+= valorTotal
      print(f"Produto: {n[0]}, preço unitário: {n[1]} quantidade em estoque:{n[2]} valor total: {valorTotal} ")
    print(f"Valor total em estoque: R${somatorio}")
  elif menu==2:
    maiorQTDEstoque=0
    nomeDoProduto=''
    for n in produtos:
      if(maiorQTDEstoque<n[1]):
        maiorQTDEstoque=n[1]
        nomeDoProduto=n[0]
    print(f"O Produto com maior quantidade em estoque é: {nomeDoProduto} com {maiorQTDEstoque} unidades em estoque")
  elif menu==3:
    menorValor=produtos[0][2]
    nomeDoProduto=''
    for n in produtos:
      if(menorValor>n[2]):
        menorValor=n[2]
        nomeDoProduto=n[0]
    print(f"O Produto mais barato em estoque é: {nomeDoProduto} com o valor de: R$ {menorValor}")
  elif menu==4:
    break
```
